Нужно написать код который будет выводить кортежи для скрытых нейронок

```python
import itertools as it
for _ in it.product([2**i for i in range(11)], repeat=4):
    print(tuple(_))
```

```python
list(it.product([2**i for i in range(11)], repeat=4))
```
